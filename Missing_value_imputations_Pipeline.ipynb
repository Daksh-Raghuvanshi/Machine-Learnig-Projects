```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
import seaborn as sns
```

For now our task is to create a pipeline for different strategies of imputations.

```python
num_vars_mean = ['LotFrontage']
num_vars_meadian = [ 'MasVnrArea', 'GarageYrBlt']
cat_vars_mode = ['Alley',
 'MasVnrType',
 'BsmtQual',
 'BsmtCond',
 'BsmtExposure',
 'BsmtFinType1',
 'BsmtFinType2',
 'Electrical',
 'FireplaceQu']
cat_vars_constant = ['GarageType',
 'GarageFinish',
 'GarageQual',
 'GarageCond',
 'PoolQC',
 'Fence',
 'MiscFeature']
```

```python
num_vars_mean_imputer = Pipeline(steps = [('imputer', SimpleImputer(strategy='mean'))])
num_vars_median_imputer = Pipeline(steps = [('imputer', SimpleImputer(strategy='median'))])
num_vars_mode_imputer = Pipeline(steps = [('imputer', SimpleImputer(strategy='most_frequent'))])
num_vars_constant_imputer = Pipeline(steps = [('imputer', SimpleImputer(strategy='constant', fill_value='missing'))])
```

```python
preprocessor = ColumnTransformer([('mean_imputer' , num_vars_mean_imputer ,num_vars_mean ),
                   ('median_imputer' ,num_vars_median_imputer, num_vars_meadian ),
                   ('mode_imputer' ,num_vars_mode_imputer,cat_vars_mode  ),
                   ('constant_imputer',num_vars_constant_imputer,  cat_vars_constant)], remainder = 'passthrough')
```

```python
preprocessor
```

```python
preprocessor.fit(X_train , )
```

```python
preprocessor.transform
```

```python
X_train_clean = preprocessor.transform(X_train)
X_test_clean = preprocessor.transform(test)
```
