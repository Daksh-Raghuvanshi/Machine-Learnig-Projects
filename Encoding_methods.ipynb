## **One Hot Encoding Using Pandas**





```python
import pandas as pd
```

```python
tips_df = pd.read_csv("/content/tips.csv")
tips_df
```

```python
dummy_df = pd.get_dummies(tips_df , drop_first= True)
```

## **One Hot Encoding using SciKit Learn**

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oh_enc = OneHotEncoder(sparse = False)
dummy_df2 = oh_enc.fit_transform(tips_df[['sex', 'smoker', 'day', 'time']])
dummy_df2
```

```python
pd.DataFrame(dummy_df2, columns = ['sex_Female', 'sex_Male', 'smoker_No',
       'smoker_Yes', 'day_Fri', 'day_Sat', 'day_Sun', 'day_Thur',
       'time_Dinner', 'time_Lunch'])
```

```python
oh_enc1 = OneHotEncoder(sparse = False, drop = 'first')
dummy_df3 = oh_enc1.fit_transform(tips_df[['sex', 'smoker', 'day', 'time']])
dummy_df3
```

```python
pd.DataFrame(dummy_df3 , columns = ['sex_Male', 'smoker_Yes', 'day_Sat',
       'day_Sun', 'day_Thur', 'time_Lunch'])
```
